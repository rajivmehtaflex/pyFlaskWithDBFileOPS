```python
currntpath=pathlib.Path.cwd()
image_path=[];
classbase=[]
for file, folder in [('moon.csv', 'moon'), ('bottle.csv', 'bottle')]:
    dest = currntpath/folder
    for i in list(dest.rglob('*.*')):
#         "image_path:{}/{},class:{}".format(i.parent.name,i.name,i.parent.name)
        image_path.append("{}/{}".format(i.parent.name,i.name))
        classbase.append("{}".format(i.parent.name))
dict = {'image_path': image_path, 'class': classbase}
df = pd.DataFrame(dict) 
df.to_csv("gd.csv")
```
